```python
img = np.zeros(shape=(512,512,3), dtype=np.uint8)
print("img의 shape: ", img.shape)

plt.subplot(1,3,1)
plt.title("all zero")
plt.imshow(img)

for i in range(0, 512):
    for j in range(0, 512):
        img[i, j, 0] =255

plt.subplot(1,3,2)
plt.title("red color")
plt.imshow(img)

for i in range(100):
    for j in range(100):
        img[i, j, 0] = 0
        img[i, j, 1] = 128
        img[i, j, 2] = 128

plt.subplot(1,3,3)
plt.title("other color")
plt.imshow(img)
```

```python
flowers_RGB = cv2.imread("./Flowers.jpg", cv2.IMREAD_COLOR)
flowers_RGB = cv2.cvtColor(flowers_RGB, cv2.COLOR_BGR2RGB)

print("data type: ", flowers_RGB.dtype)
print("shape: ", flowers_RGB.shape)
print("flowers_RGB[0,0]값: ", flowers_RGB[0,0])

plt.imshow(flowers_RGB)
```

```python
flower_RGB = cv2.imread("./Flowers.jpg", cv2.IMREAD_COLOR)
flower_RGB = cv2.cvtColor(flower_RGB, cv2.COLOR_BGR2RGB)

flower_R = flower_RGB[:,:,0]
flower_G = flower_RGB[:,:,1]
flower_B = flower_RGB[:,:,2]

print("flowers_RGB의 shape : ", flower_RGB.shape)
print("flowers_RGB에서 R채널의 shpae", flower_R.shape)

plt.figure()
plt.subplot(1,3,1)
plt.title("R channel")
plt.imshow(flower_R, cmap='gray')

plt.figure()
plt.subplot(1,3,2)
plt.title("G channel")
plt.imshow(flower_G, cmap='gray')

plt.figure()
plt.subplot(1,3,3)
plt.title("B channel")
plt.imshow(flower_B, cmap='gray')
```

```python
lena_gray = cv2.imread("./Lena_gray.jpg", cv2.IMREAD_GRAYSCALE)

plt.subplot(1,3,1)
plt.title("original")
plt.imshow(lena_gray, cmap='gray')

h, w = lena_gray.shape
lena_add = np.zeros((h,w), np.uint8)

for i in range(h):
    for j in range(w):
        lena_add[i, j] = lena_gray[i, j] + 50
plt.subplot(1,3,2)
plt.title("Add 50")
plt.imshow(lena_add, cmap="gray")

lena_add = lena_gray+100

plt.subplot(1,3,3)
plt.title("Add 100")
plt.imshow(lena_add, cmap="gray")
```

```python
lena_div = lena_gray / 2

plt.figure()
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(lena_gray, cmap='gray')

plt.subplot(1,2,2)
plt.title("Div")
plt.imshow(lena_div, cmap='gray', vmin=0, vmax=255)
```

```python
lena_gray = cv2.imread("./Lena_gray.jpg", cv2.IMREAD_GRAYSCALE)

plt.subplot(1,3,1)
plt.title("original image")
plt.imshow(lena_gray, cmap='gray')

lena_mul = np.int32(lena_gray) * 2
lena_mul[lena_mul > 255] = 255
lena_mul =np.uint8(lena_mul)

plt.subplot(1,3,2)
plt.title("mul (saturated)")
plt.imshow(lena_mul, cmap='gray')

lena_sub = np.int32(lena_gray)-100
lena_sub[lena_sub<0]=0
lena_sub =np.uint8(lena_sub)

plt.subplot(1,3,3)
plt.title("sub (saturated)")
plt.imshow(lena_sub, cmap='gray')
```

```python
lena = cv2.imread("./lena_color.jpg")
lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)

pattern = cv2.imread("./Pattern1.png")
pattern = cv2.cvtColor(pattern, cv2.COLOR_BGR2RGB)

alpha = 0.5
beta = 1-alpha
mixed = (alpha*lena) + (beta*pattern)
mixed = np.uint8(mixed)

plt.subplot(1,3,1)
plt.imshow(lena)
plt.subplot(1,3,2)
plt.imshow(pattern)
plt.subplot(1,3,3)
plt.imshow(mixed)
```

```python
img1 = cv2.imread("./Pattern1.png", 0)
img2 = cv2.imread("./Pattern2.png", 0)

img_diff = np.abs(img1 - img2)

plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)

plt.figure()
plt.imshow(img_diff, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
```
